## Modelo base
Replica del modelo de Dias, utilizado unicamente como baseline de las pruebas.
Lamentablemente el modelo no puede estar completo al no saber la dimensionalidad de los datos de entrada por lo que su validez como replica no puede ser aceptada.

```python
class Dias_Model(Module):
    
    def __init__(self):
        super(Dias_Model, self).__init__()
        
        # Capa 1
        self.conv1 = Conv1d()
        self.batchnorm1 = BatchNorm1d()
        self.relu1 = ReLU()
    
        # Capa 2
        self.conv2 = Conv1d()
        self.relu2 = ReLU()
        self.droput1 = Dropout()
        self.batchnorm2 = BatchNorm1d()
    
        # Maxpool
        self.maxpool = Maxpool1D()

        # Intermedias
        self.intermedia_conv1 = Conv1d()
        self.intermeida_relu1 = ReLU()
    
        self.intermedia_conv2 = Conv1d()
        self.intermeida_relu2 = ReLU()
    
        self.intermedia_conv3 = Conv1d()
        self.intermedia_batchnorm = BatchNorm1d()
        self.intermeida_relu3 = ReLU()
        self.intermedia_dropout = Dropout()
    
        # Final
        self.final_conv = Conv1d()
        self.final_view = View()
        self.final_dropout = Dropout()

        # Clasificador
        self.final_clasificador = Linear()
        self.final.batchnorm = BatchNorm1D()
        self.final_softmax = Softmax()
    
        
        
        
    def forward(self, x):
        # Capa 1
        x = self.relu1(self.batchnorm1(self.conv1(x))) 
        
        # Capa 2
        x = self.batchnorm2(self.droput1(self.relu2(self.conv2(x))))
        
        # Maxpool
        x = self.maxpool(x)
        
        # Intermedias 1
        x = self.intermeida_relu1(self.intermedia_conv1(x))    
        
        # Intermedias 2
        x = self.intermeida_relu2(self.intermedia_conv2(x))
    
        # Intermedias 3
        x = self.intermedia_dropout(self.intermeida_relu3(self.intermedia_batchnorm(self.intermedia_conv3(x))))
        
        # Final
        x = self.final_dropout(self.final_view(self.final_conv(x)))
        
        # Clasificador
        x = self.final.batchnorm(self.final_clasificador(x))
        output = self.final_softmax(x)
        
        return output   
```
